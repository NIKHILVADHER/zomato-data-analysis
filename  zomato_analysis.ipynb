```python
df=df.drop(['url', 'address','phone', 'reviews_list', 'menu_item'],axis=1)
```

```python
df=df.dropna()
```

```python
df.shape
df['rate'].unique()[:20]
```

```python
df=df[df["rate"]!="NEW"]
df=df[df["rate"]!="-"]
```

```python
df["rate"]=df["rate"].str.replace("/5","")
```

```python
df["rate"]=df["rate"].str.strip()
```

```python
df['rate'].unique()[:20]
```

```python
df["rate"]=df["rate"].astype(float)
```

```python
df["rate"].head()
```

```python
df["rate"].describe()
```

```python

df['rate'].dtype
```

```python
import matplotlib.pyplot as plt
plt.hist(df["rate"],bins=20)
plt.xlabel("Ratings")
plt.ylabel("Number of Restaurants")
plt.title("Distribution of Restaurant Ratings")
plt.show()
```

Insight:

The majority of restaurants have ratings between 3.5 and 4, indicating that most restaurants maintain moderately good customer satisfaction.

```python
df["online_order"].value_counts()
```

Insight:

The majority of restaurants offer online ordering.
Around 65% of restaurants provide online delivery services, showing that online food ordering is widely adopted.

```python
import matplotlib.pyplot as plt
df["online_order"].value_counts().plot(kind="bar")
plt.xlabel("Online Order Availability")
plt.ylabel("NUmber of restaurant")
plt.title("Online food ordering availability")
plt.show()
```

```python
df["book_table"].value_counts()
```

```python
import matplotlib.pyplot as plt
df["book_table"].value_counts().plot(kind="bar")
plt.xlabel("Booking Availability")
plt.ylabel("Number of restaurants")
plt.title("Table Booking Availability Check")
plt.show()
```

Insight:

Most restaurants do not provide table booking facilities. Only around 15% of restaurants allow reservations, suggesting that table booking is mainly available in premium or fine-dining restaurants.

```python
df.groupby("book_table")["rate"].mean()
```

Insight:

Restaurants that allow table booking have a higher average rating (4.16) compared to restaurants without table booking (3.81). This suggests that reservation-enabled restaurants are typically higher-quality or premium dining establishments.

```python
df.groupby("book_table")["rate"].mean().plot(kind="bar")

plt.xlabel("Table Booking")
plt.ylabel("Average Rating")
plt.title("Table Booking vs Average Rating")

plt.show()
```
